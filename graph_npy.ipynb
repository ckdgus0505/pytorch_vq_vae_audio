```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tr_recon_error = np.load('training_reconstruction_errors_per_epochs.npy')
te_recon_error = np.load('validation_reconstruction_errors_per_epochs.npy')
tr_total_error = np.load('training_total_loss_per_epochs.npy')
te_total_error = np.load('validation_total_loss_per_epochs.npy')
tr_mcd = np.load('mean_of_mcd_per_epochs.npy')
te_mcd = np.load('validation_mean_of_mcd_per_epochs.npy')
```

```python
print(len(tr_recon_error))
print(len(te_recon_error))
print(len(tr_total_error))
print(len(te_total_error))
# print(len(tr_mcd))
# print(len(te_mcd))
```

```python
xax = np.arange(20,501, 20)
```

```python
plt.subplot(3,1,1)
plt.ylabel('total_loss')

plt.plot(te_total_error, label = 'validation', color='darkorange', linewidth=1)
plt.plot(tr_total_error, label = 'train', color='steelblue', linewidth=1)
plt.text(np.argmin(tr_total_error), np.min(tr_total_error)+1.5, s=('('+str(np.argmin(tr_total_error)+1)+","+str(np.min(tr_total_error))[:6])+')')
plt.plot(np.argmin(tr_total_error),np.min(tr_total_error), 'xb')

# plt.text(np.argmin(tr_total_error[100:]+100), np.min(tr_total_error[100:]) + 0.19, s=('('+str(np.argmin(tr_total_error[100:]+100)+1)+","+str(np.min(tr_total_error[100:]))[:6])+')')
# plt.plot(np.argmin(tr_total_error[100:]+100),np.min(tr_total_error[100:]), 'xb')



plt.text(np.argmin(te_total_error), np.min(te_total_error), s=('('+str(np.argmin(te_total_error)+1)+","+str(np.min(te_total_error))[:6])+')')
plt.plot(np.argmin(te_total_error),np.min(te_total_error), 'xr')
plt.legend()

plt.subplot(3,1,2)
plt.ylabel('reconstruction_loss')
plt.plot(te_recon_error, label ='validation', color='darkorange', linewidth=1)
plt.plot(tr_recon_error, label = 'train', color='steelblue', linewidth=1)
#plt.plot(np.argmin(tr_recon_error), np.min(tr_recon_error), 'xr'  )
plt.text(np.argmin(tr_recon_error), np.min(tr_recon_error)+0.0001, s=('('+str(np.argmin(tr_recon_error)+1)+","+str(np.min(tr_recon_error))[:6])+')')
plt.plot(np.argmin(tr_recon_error),np.min(tr_recon_error), 'xb')

#plt.text(np.argmin(tr_recon_error[100:])+100, np.min(tr_recon_error[100:]), s=('('+str(np.argmin(tr_recon_error[100:])+101)+","+str(np.min(tr_recon_error[100:]))[:6])+')')
#plt.plot(np.argmin(tr_recon_error[100:])+100,np.min(tr_recon_error[100:]), 'xb')

plt.text(np.argmin(te_recon_error), np.min(te_recon_error), s=('('+str(np.argmin(te_recon_error)+1)+","+str(np.min(te_recon_error))[:6])+')')
plt.plot(np.argmin(te_recon_error),np.min(te_recon_error), 'xr')
plt.legend()

plt.subplot(3,1,3)
plt.ylabel('mcd')
plt.plot(xax, tr_mcd, label = 'train')
plt.plot(xax, te_mcd, label = 'validation')
plt.text(np.argmin(tr_mcd)*20+20, np.min(tr_mcd) - 0.5, s=('('+str(np.argmin(tr_mcd)*20+20)+","+str(np.min(tr_mcd))[:6])+')')
plt.plot(np.argmin(tr_mcd)*20+20,np.min(tr_mcd), 'xb')

plt.text(np.argmin(te_mcd)*20+20, np.min(te_mcd), s=('('+str(np.argmin(te_mcd)*20+20)+","+str(np.min(te_mcd))[:6])+')')
plt.plot(np.argmin(te_mcd)*20+20,np.min(te_mcd), 'xr')
plt.legend()

plt.savefig('losses.png', dpi=300)
```

```python
lrs = np.load('lrs.npy')
plt.plot(lrs, label='learning rate')
plt.plot(np.argmax(lrs)+1,np.max(lrs), 'ok')
plt.plot(np.argmin(lrs)+1,np.min(lrs), 'ok')
plt.text(np.argmax(lrs)+4, np.max(lrs), s=('('+str(np.argmax(lrs)+1)+","+str(np.max(lrs))[:6])+')')
plt.text(np.argmin(lrs)-40, np.min(lrs)+0.00000002, s=('('+str(np.argmin(lrs)+1)+","+str( round(lrs[-1],8) ) )+')')
plt.title('learning rate decay')
plt.ylabel('learning rate')
plt.xlabel('epoch')
plt.legend()
# plt.savefig('learning_rate.png', dpi=300)
```
