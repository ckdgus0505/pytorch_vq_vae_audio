```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tr_recon_error = np.load('training_reconstruction_errors_per_epochs300.npy')
te_recon_error = np.load('validation_reconstruction_errors_per_epochs300.npy')
tr_total_error = np.load('training_total_loss_per_epochs300.npy')
te_total_error = np.load('validation_total_loss_per_epochs300.npy')
tr_mcd = np.load('training_mcd_per_epochs.npy')
te_mcd = np.load('test_mcd_per_epochs.npy')
```

```python
print(len(tr_recon_error))
print(len(te_recon_error))
print(len(tr_total_error))
print(len(te_total_error))
```

```python
xax = np.arange(10,301, 10)
```

```python
plt.subplot(3,1,1)
plt.ylabel('total_loss')
plt.plot(tr_total_error, label = 'train')
plt.plot(te_total_error, label = 'validation')
plt.text(np.argmin(tr_total_error+1), np.min(tr_total_error) + 0.03, s=('('+str(np.argmin(tr_total_error)+1)+","+str(np.min(tr_total_error))[:6])+')')
plt.plot(np.argmin(tr_total_error+1),np.min(tr_total_error), 'xb')

plt.text(np.argmin(te_total_error+1), np.min(te_total_error), s=('('+str(np.argmin(te_total_error)+1)+","+str(np.min(te_total_error))[:6])+')')
plt.plot(np.argmin(te_total_error+1),np.min(te_total_error), 'xr')
plt.legend()

plt.subplot(3,1,2)
plt.ylabel('reconstruction_loss')
plt.plot(tr_recon_error, label = 'train')
plt.plot(te_recon_error, label ='validation')
#plt.plot(np.argmin(tr_recon_error), np.min(tr_recon_error), 'xr'  )
plt.text(np.argmin(tr_recon_error+1), np.min(tr_recon_error), s=('('+str(np.argmin(tr_recon_error)+1)+","+str(np.min(tr_recon_error))[:6])+')')
plt.plot(np.argmin(tr_recon_error+1),np.min(tr_recon_error), 'xb')

plt.text(np.argmin(te_recon_error+1), np.min(te_recon_error), s=('('+str(np.argmin(te_recon_error)+1)+","+str(np.min(te_recon_error))[:6])+')')
plt.plot(np.argmin(te_recon_error+1),np.min(te_recon_error), 'xr')
plt.legend()

plt.subplot(3,1,3)
plt.ylabel('mcd')
plt.plot(xax, tr_mcd, label = 'train')
plt.plot(xax, te_mcd, label = 'validation')
plt.text(np.argmin(tr_mcd)*10+10, np.min(tr_mcd) + 0.03, s=('('+str(np.argmin(tr_mcd)*10+10)+","+str(np.min(tr_mcd))[:6])+')')
plt.plot(np.argmin(tr_mcd)*10+10,np.min(tr_mcd), 'xb')

plt.text(np.argmin(te_mcd)*10+10, np.min(te_mcd), s=('('+str(np.argmin(te_mcd)*10+10)+","+str(np.min(te_mcd))[:6])+')')
plt.plot(np.argmin(te_mcd)*10+10,np.min(te_mcd), 'xr')
plt.legend()

plt.savefig('losses.png', dpi=300)
```
